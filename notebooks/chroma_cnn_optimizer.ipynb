```python
import utils
import models
import model_utils
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.optim as optim

plt.style.use('dark_background')
```

```python
dl = model_utils.DataLoader("chroma_cqt", cuda=True)
```

```python
chroma_cnn = models.ChromaCNNFreq(add_name="_adam").cuda()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(chroma_cnn.parameters(), lr=0.001)
mutils = model_utils.ModelUtils(chroma_cnn, criterion, optimizer, dl)
```

```python
mutils.train(n_epochs=100)
```

```python
mutils.get_accuracies_from_pths(n_epochs=100, cuda=True)

plt.figure(figsize=(10, 4))
plt.suptitle("Single feature map, Adam Optimizer")
plt.subplot(1, 2, 1)
plt.title("Loss")
plt.plot(mutils.load_losses())
plt.xticks([])
plt.ylabel("Loss")
plt.subplot(1, 2, 2)
plt.title("Accuracy")
mutils.plot_accuracies()
```

```python
chroma_cnn = models.ChromaCNNFreq(add_name="_sgd").cuda()
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(chroma_cnn.parameters(), lr=0.01, momentum=0.9)
mutils = model_utils.ModelUtils(chroma_cnn, criterion, optimizer, dl)
```

```python
mutils.train(n_epochs=100)
```

```python
mutils.get_accuracies_from_pths(n_epochs=100, cuda=True)

plt.figure(figsize=(10, 4))
plt.suptitle("Single feature map, SGD Optimizer")
plt.subplot(1, 2, 1)
plt.title("Loss")
plt.plot(mutils.load_losses())
plt.xticks([])
plt.ylabel("Loss")
plt.subplot(1, 2, 2)
plt.title("Accuracy")
mutils.plot_accuracies()
```
