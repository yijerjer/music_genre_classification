
## Understanding Features

* **Chroma** refers to the twelve semitones/pitch classes in an octave
* **Constant-Q Transform (CQT)** is a type of Discrete Fourier Transform that produces a logarithmic frequency space. It is more suitable in music as it replicates a human's auditory system, where the frequency resolution at high frequencies is lower than at low frequencies.
* `chroma_cqt` obtains the (average or sum?) amplitude of a particular chroma across all octaves at a particular time window. `chroma_cens` is a function of `chroma_cqt`, and is described to be the *Energy Normalised* chroma.
* `chroma_stft` calculates the **Short Time Fourier Transform (STFT)** of the track, and bin them into chromas. (Note that binning into chromas creates a chromagram)
* `mfcc` is the **Mel-frequency Cepstral Coefficients**. This coefficients are obtained through the following steps: Take the FT of a windowed signal, 
 convert linear into mel scale and take the log, take the consine transform of the log powers, the amplitudes are the coefficients.
* `tonnetz` is described to be a 2d lattice diagram that represents tonal space, that is used to visualise chords and harmonic relationships. Assume that in this case, this values returned are a way to quantify this.
* `zcr` (zero crossing rate) is the number of times the time series data crosses zero.
* `rmse` is the root mean square of the energy, calculated through the STFT.
* `spectral_bandwidth`, `spectral_centroid`, `spectral_contrast`, `spectral_rolloff` all relates to the frequency spectrum obtained from a STFT.


### Plotting the Features of a Track

```python
import librosa, librosa.display
import IPython.display as ipd
import numpy as np

small_tracks = utils.load_csv("tracks")
```

```python
random_sample = small_tracks.sample()
sample_id = random_sample.index[0]
sample_genre = random_sample["track", "genre_top"].item()
sample_waveform, sr = utils.load_waveform(sample_id, sr=22050)
print(f"Track ID: {sample_id}, Genre: {sample_genre}")

librosa.display.waveplot(sample_waveform)
plt.show()
```

```python
ipd.Audio(data=sample_waveform, rate=sr)
```

#### CQT, Chroma CQT, Chroma CENS

```python
sample_cqt = np.abs(librosa.cqt(sample_waveform, sr=sr, hop_length=1024, bins_per_octave=12, n_bins=7*12))
chroma_cqt = librosa.feature.chroma_cqt(C=sample_cqt, hop_length=1024)
chroma_cens = librosa.feature.chroma_cens(C=sample_cqt, hop_length=1024, sr=sr)
print(f"Shape of CQT: {sample_cqt.shape}, Shape of Chroma CQT: {chroma_cqt.shape}, Shape of Chroma CENS: {chroma_cens.shape}")

plt.figure(figsize=(10, 15))
plt.subplot(3, 1, 1)
librosa.display.specshow(sample_cqt, y_axis="cqt", x_axis="time", sr=sr, hop_length=1024)
plt.title("CQT")
plt.subplot(3, 1, 2)
librosa.display.specshow(chroma_cqt, y_axis="chroma", x_axis="time", sr=sr, hop_length=1024)
plt.title("Chroma CQT")
plt.subplot(3, 1, 3)
librosa.display.specshow(chroma_cens, y_axis="chroma", x_axis="time", sr=sr, hop_length=1024)
plt.title("Chroma CENS")

plt.show()
```

#### STFT, Mel STFT, Chroma STFT

```python
stft_linear = np.abs(librosa.stft(sample_waveform, n_fft=1024, hop_length=1024))
stft_linear_db = librosa.amplitude_to_db(stft_linear, ref=np.max)
stft_mel = librosa.feature.melspectrogram(S=stft_linear**2, n_fft=1024, hop_length=1024)
stft_mel_db = librosa.amplitude_to_db(stft_mel, ref=np.max)
chroma_stft = librosa.feature.chroma_stft(S=stft_linear**2, hop_length=1024, n_fft=1024, n_chroma=12)

plt.figure(figsize=(10, 15))
plt.subplot(3, 1, 1)
librosa.display.specshow(stft_linear_db, y_axis="linear", x_axis="time", sr=sr, hop_length=1024)
plt.title("STFT")
plt.subplot(3, 1, 2)
librosa.display.specshow(stft_mel_db, y_axis="mel", x_axis="time", sr=sr, hop_length=1024)
plt.title("Mel STFT")
plt.subplot(3, 1, 3)
librosa.display.specshow(chroma_stft, y_axis="chroma", x_axis="time", sr=sr, hop_length=1024)
plt.title("Chroma STFT")

plt.show()
```

#### Tonnetz, MFCC

```python
tonnetz = librosa.feature.tonnetz(chroma=chroma_cens)
mfcc = librosa.feature.mfcc(S=stft_mel_db, n_mfcc=20)

plt.figure(figsize=(10, 15))
plt.subplot(3, 1, 1)
librosa.display.specshow(tonnetz, sr=sr, x_axis='time', y_axis='tonnetz')
plt.title("Tonnetz")
plt.subplot(3, 1, 2)
librosa.display.specshow(mfcc, sr=sr, x_axis='time')
plt.title("MFCC")

plt.show()
```

#### Spectral Centroid, Bandwidth, Rolloff, Contrast, Zero Crossing Rate

```python
from sklearn import preprocessing

def normalize(x):
    return preprocessing.minmax_scale(x)

centroid = librosa.feature.spectral_centroid(sample_waveform, sr=sr, n_fft=1024, hop_length=1024)[0]
bandwidth = librosa.feature.spectral_bandwidth(sample_waveform, sr=sr, n_fft=1024, hop_length=1024)[0]
rolloff = librosa.feature.spectral_rolloff(sample_waveform, sr=sr, n_fft=1024, hop_length=1024)[0]
contrast = librosa.feature.spectral_contrast(sample_waveform, sr=sr, n_fft=1024, hop_length=1024)
zcr = librosa.feature.zero_crossing_rate(sample_waveform, hop_length=1024)[0]

plt.figure(figsize=(10, 25))
plt.subplot(5, 1, 1)
librosa.display.waveplot(sample_waveform, alpha=0.5)
plt.plot(np.linspace(0, 30, len(centroid)), normalize(centroid))
plt.title("Spectral Centroid")
plt.subplot(5, 1, 2)
librosa.display.waveplot(sample_waveform, alpha=0.5)
plt.plot(np.linspace(0, 30, len(bandwidth)), normalize(bandwidth))
plt.title("Spectral Bandwidth")
plt.subplot(5, 1, 3)
librosa.display.waveplot(sample_waveform, alpha=0.5)
plt.plot(np.linspace(0, 30, len(rolloff)), normalize(rolloff))
plt.title("Spectral Rolloff")
plt.subplot(5, 1, 4)
librosa.display.specshow(contrast, x_axis="time", sr=sr, hop_length=1024)
plt.title("Spectral Contrast")
plt.subplot(5, 1, 5)
librosa.display.waveplot(sample_waveform, alpha=0.5)
plt.plot(np.linspace(0, 30, len(rolloff)), normalize(zcr))
plt.title("Zero Crossing Rate")

plt.show()
```
